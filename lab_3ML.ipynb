```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("canada_per_capita_income.csv")

# Rename columns for easier reference
df.columns = ["year", "income"]

# Prepare data for regression model
X = df[["year"]]  # Feature: Year
y = df["income"]   # Target: Per capita income

# Train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predict per capita income for 2020
predicted_income = model.predict([[2020]])
print("Predicted per capita income for 2020:", predicted_income[0])

# Visualization
plt.scatter(X, y, color='blue', label="Actual Data")  # Plot actual data points
plt.plot(X, model.predict(X), color='red', linewidth=2, label="Regression Line")  # Plot regression line
plt.scatter([[2020]], predicted_income, color='green', marker='o', s=100, label="Prediction (2020)")  # Highlight prediction

# Labels and title
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.title("Canada Per Capita Income Prediction")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
salary_df = pd.read_csv("salary.csv")

# Check for missing values and drop them
salary_df_cleaned = salary_df.dropna(subset=["YearsExperience"])

# Prepare data for regression model
X_salary = salary_df_cleaned[["YearsExperience"]]  # Feature: Years of Experience
y_salary = salary_df_cleaned["Salary"]             # Target: Salary

# Train the Linear Regression model
salary_model = LinearRegression()
salary_model.fit(X_salary, y_salary)

# Predict salary for 12 years of experience
predicted_salary = salary_model.predict([[12]])
print("Predicted salary for 12 years of experience:", predicted_salary[0])

# Visualization
plt.scatter(X_salary, y_salary, color='blue', label="Actual Data")  # Plot actual data points
plt.plot(X_salary, salary_model.predict(X_salary), color='red', linewidth=2, label="Regression Line")  # Plot regression line
plt.scatter([[12]], predicted_salary, color='green', marker='o', s=100, label="Prediction (12 Years Exp)")  # Highlight prediction

# Labels and title
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary Prediction Based on Experience")
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
hiring_df = pd.read_csv("hiring.csv")

# Function to convert experience words to numbers
def convert_experience(value):
    word_to_number = {
        "zero": 0, "one": 1, "two": 2, "three": 3, "four": 4, "five": 5,
        "six": 6, "seven": 7, "eight": 8, "nine": 9, "ten": 10, "eleven": 11, "twelve": 12
    }
    return word_to_number.get(value.lower(), None) if isinstance(value, str) else value

# Convert experience column
hiring_df["experience"] = hiring_df["experience"].apply(convert_experience)

# Fill missing values
hiring_df["experience"].fillna(hiring_df["experience"].median(), inplace=True)
hiring_df["test_score(out of 10)"].fillna(hiring_df["test_score(out of 10)"].mean(), inplace=True)

# Prepare data for training
X_hiring = hiring_df[["experience", "test_score(out of 10)", "interview_score(out of 10)"]]
y_hiring = hiring_df["salary($)"]

# Train the regression model
hiring_model = LinearRegression()
hiring_model.fit(X_hiring, y_hiring)

# Predict salaries for given candidates
candidates = [[2, 9, 6], [12, 10, 10]]
predicted_salaries = hiring_model.predict(candidates)

# Display results
print("Predicted salary for 2 years experience, 9 test score, 6 interview score:", predicted_salaries[0])
print("Predicted salary for 12 years experience, 10 test score, 10 interview score:", predicted_salaries[1])
```
